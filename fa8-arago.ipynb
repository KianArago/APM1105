1. Create an exponent calculator using both iterations and recursion, with rendering time.

```python
import time 

  

def iterative_exponent(base, exponent): 

    result = 1 

    for _ in range(exponent): 

        result *= base 

    return result 

  

def recursive_exponent(base, exponent): 

    if exponent == 0: 

        return 1 

    elif exponent == 1: 

        return base 

    else: 

        return base * recursive_exponent(base, exponent - 1) 

  

def calculate_and_display(base, exponent): 

    start_time = time.time() 

    iterative_result = iterative_exponent(base, exponent) 

    end_time = time.time() 

    iterative_time = end_time - start_time 

  

    start_time = time.time() 

    recursive_result = recursive_exponent(base, exponent) 

    end_time = time.time() 

    recursive_time = end_time - start_time 

  

    print(f"Iterative Result: {iterative_result}, Rendering Time: {iterative_time:.8f} seconds") 

    print(f"Recursive Result: {recursive_result}, Rendering Time: {recursive_time:.8f} seconds") 

  



base = int(input("Enter the base: ")) 

exponent = int(input("Enter the exponent: ")) 

calculate_and_display(base, exponent) 
```

Which of the two functions performs faster? And why is it performing faster?
- Recursion performs better because recursion is concise and simpler in coding, although recursion is faster in the test, iterative might be better for more complex, larger numbers and codes. In other test, given larger bases and exponents, iteration performs better than recursion.

2. Iteratively and recursively calculate the sum of the first n powers of 2, along with the rendering time of each function using the time module.

```python
import time 

  

def iterative_sum_of_powers(n): 

    result = 0 

    for i in range(n): 

        result += 2 ** i 

    return result 

  

def recursive_sum_of_powers(n): 

    if n == 0: 

        return 0 

    else: 

        return (2 ** (n - 1)) + recursive_sum_of_powers(n - 1) 

  

def calculate_and_display_sum(n): 

    start_time = time.time() 

    iterative_result = iterative_sum_of_powers(n) 

    end_time = time.time() 

    iterative_time = end_time - start_time 

  

    start_time = time.time() 

    recursive_result = recursive_sum_of_powers(n) 

    end_time = time.time() 

    recursive_time = end_time - start_time 

  

    print(f"Iterative Sum: {iterative_result}, Rendering Time: {iterative_time:.8f} seconds") 

    print(f"Recursive Sum: {recursive_result}, Rendering Time: {recursive_time:.8f} seconds") 

  



n = int(input("Enter the value of n: ")) 

calculate_and_display_sum(n) 
```

Which of the two functions performs faster? And why is it performing faster?
- In this test, iteration performs faster and better, because while iteration performs faster when given larger numbers, recursion performs better when numbers are small.
